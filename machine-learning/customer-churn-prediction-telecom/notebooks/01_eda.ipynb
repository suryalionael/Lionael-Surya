# 01 Exploratory Data Analysis (EDA)
This notebook loads and analyzes the raw telecom churn dataset to uncover demographics, account, and service trends correlated with customer churn.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load configurations
from src.config import CONFIG
from src.preprocessing import clean_data

df_raw = pd.read_csv(CONFIG["paths"]["raw_data"])
df = clean_data(df_raw, is_train=True)
df.head()
```

### Target Imbalance Analysis

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df_raw, palette='Blues')
plt.title('Target Distribution (Churn Status)')
plt.show()
```

### Demographics vs. Churn

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.countplot(x='Contract', hue='Churn', data=df_raw, ax=axes[0], palette='Set1')
axes[0].set_title('Churn by Contract Type')

sns.countplot(x='PaymentMethod', hue='Churn', data=df_raw, ax=axes[1], palette='Set2')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=30)
axes[1].set_title('Churn by Payment Method')
plt.tight_layout()
plt.show()
```
